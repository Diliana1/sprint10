## AVG a Minuts

```python
# Función para convertir un tiempo en formato "min:seg/km" a minutos decimales
def convertir_avg_a_minutos(avg):
    # Asegurarse de que 'avg' es una cadena
    avg = str(avg)
    
    # Eliminar cualquier texto adicional como '/km'
    avg = avg.replace('/km', '').strip()

    # Verificar si el formato es adecuado, para evitar errores en conversiones mal formadas
    if "'" in avg:  # Si tiene el formato min:seg
        minutos, segundos = avg.split("'")
        segundos = segundos.replace('"', '')  # Eliminar comillas del string de segundos
        # Convertir a minutos decimales
        minutos_decimales = int(minutos) + int(segundos) / 60
        return minutos_decimales
    else:
        return None  # O retornar un valor que indique que no tiene el formato esperado

# Aplicar la conversión a la columna 'AVG'
df_concatenado.loc[:, 'AVG_minutos'] = df_concatenado['AVG'].apply(convertir_avg_a_minutos)






```

```python
from scipy.stats import anderson

# Columnas a evaluar
columns_to_test = ['Time_minutes', 'GAP_minutes', 'AVG_minutos']

# Realizamos la prueba de Anderson-Darling para cada columna
for col in columns_to_test:
    print(f"\nPrueba de Anderson-Darling para la columna: {col}")
    
    # Eliminamos valores nulos antes de realizar la prueba
    sample = df_concatenado[col].dropna().sample(5000, random_state=42)  # Tomamos una muestra si es necesario
    
    # Aplicamos la prueba de Anderson-Darling
    result = anderson(sample, dist='norm')
    print(f"  Estadístico A-D: {result.statistic:.3f}")
    
    # Los valores críticos para el nivel de significancia (dependiendo del valor p)
    for i in range(len(result.critical_values)):
        sl = result.significance_level[i]
        cv = result.critical_values[i]
        print(f"  Nivel de significancia {sl}%: {cv:.3f} (Valor crítico)")
    
    # Evaluamos la normalidad
    if result.statistic > result.critical_values[2]:  # Comparando con el nivel de 5%
        print(" Rechaza la hipotesis.")
    else:
        print("  Reachaza La muestra  no sigue una distribución normal.")
```

```python
from scipy.stats import anderson

# Filtramos los datos por sexo
sexo_groups = df_concatenado.groupby('Sexo')

# Seleccionamos una columna para probar (por ejemplo, Time_minutes)
column_to_test = 'Time_minutes'

# Limitar el tamaño de la muestra a 1000 filas para cada grupo
sample_size = 1000

for sexo, group in sexo_groups:
    print(f"\nPrueba de Anderson-Darling para el sexo: {sexo}")
    
    # Seleccionamos una muestra aleatoria si el grupo es más grande que sample_size
    sample = group[column_to_test].dropna().sample(min(len(group), sample_size), random_state=42)
    
    # Aplicamos la prueba de Anderson-Darling
    result = anderson(sample, dist='norm')
    print(f"  Estadístico A-D: {result.statistic:.3f}")
    
    # Los valores críticos para el nivel de significancia
    for i in range(len(result.critical_values)):
        sl = result.significance_level[i]
        cv = result.critical_values[i]
        print(f"  Nivel de significancia {sl}%: {cv:.3f} (Valor crítico)")
    
    # Evaluamos la normalidad
    if result.statistic > result.critical_values[2]:  # Comparando con el nivel de 5%
        print(" rechaza La muestra no  sigue una distribución normal.")
    else:
        print("   rechazaLa muestra sigue una distribución normal.")
```
